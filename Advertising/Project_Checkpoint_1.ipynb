```python
#SMOTE
smote = SMOTE()
ad_x_smote, ad_y_smote = smote.fit_resample(standardized_ad_features, ad_target)
```

```python
#Visualize Classes After Smote
plt.figure(figsize=(6, 6))
sns.countplot(x='click', data=pd.DataFrame(ad_y_smote, columns=['click']), hue='click') 
plt.show()
```
